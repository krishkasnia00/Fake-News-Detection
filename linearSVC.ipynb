```python
import joblib
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
def train_model(X_train, X_test, y_train, y_test):
    model = LinearSVC()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return accuracy_score(y_test, y_pred)
```

```python
def evaluate_dataset(name):
    # Load labels
    y_train = joblib.load(f"{name}_y_train.pkl")
    y_test = joblib.load(f"{name}_y_test.pkl")
    
    # ===== TF-IDF =====
    X_train_tfidf = joblib.load(f"{name}_X_train_tfidf.pkl")
    X_test_tfidf = joblib.load(f"{name}_X_test_tfidf.pkl")
    
    model_tfidf = LinearSVC()
    model_tfidf.fit(X_train_tfidf, y_train)
    
    y_pred_tfidf = model_tfidf.predict(X_test_tfidf)
    acc_tfidf = accuracy_score(y_test, y_pred_tfidf)
    
    # ✅ SAVE TF-IDF MODEL
    joblib.dump(model_tfidf, f"{name}_tfidf_model.pkl")
    
    
    # ===== BoW =====
    X_train_bow = joblib.load(f"{name}_X_train_bow.pkl")
    X_test_bow = joblib.load(f"{name}_X_test_bow.pkl")
    
    model_bow = LinearSVC()
    model_bow.fit(X_train_bow, y_train)
    
    y_pred_bow = model_bow.predict(X_test_bow)
    acc_bow = accuracy_score(y_test, y_pred_bow)
    
    # ✅ SAVE BoW MODEL
    joblib.dump(model_bow, f"{name}_bow_model.pkl")
    
    
    # ===== PRINT =====
    print(f"\n{name.upper()} Results:")
    print("TF-IDF Accuracy:", acc_tfidf)
    print("BoW Accuracy:", acc_bow)
    
    return acc_tfidf, acc_bow
```

```python
# Run for both datasets
evaluate_dataset("isot")
evaluate_dataset("welfake")
```
