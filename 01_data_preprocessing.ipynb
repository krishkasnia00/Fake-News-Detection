```python
# Basic Libraries
import pandas as pd
import numpy as np
import re
import os

# NLP Libraries
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download stopwords (run once)
nltk.download('stopwords')
```

```python
# Define base path (change if needed)
BASE_PATH = "../Data"

ISOT_FAKE_PATH = os.path.join(BASE_PATH, "ISOT", r"C:\Users\Krish Kasnia\Desktop\FAKE NEWS DETECTION\Data\ISOT\Fake.csv")
ISOT_TRUE_PATH = os.path.join(BASE_PATH, "ISOT", r"C:\Users\Krish Kasnia\Desktop\FAKE NEWS DETECTION\Data\ISOT\True.csv")
WELFAKE_PATH = os.path.join(BASE_PATH, "WELFake", r"C:\Users\Krish Kasnia\Desktop\FAKE NEWS DETECTION\Data\WELFake\WELFake_Dataset.csv")
```

```python
# Load ISOT datasets
isot_fake = pd.read_csv(ISOT_FAKE_PATH)
isot_true = pd.read_csv(ISOT_TRUE_PATH)

# Load WELFake dataset
welfake = pd.read_csv(WELFAKE_PATH)

print("ISOT Fake Shape:", isot_fake.shape)
print("ISOT True Shape:", isot_true.shape)
print("WELFake Shape:", welfake.shape)
```

```python
# ISOT labels
isot_fake["label"] = 1   # Fake
isot_true["label"] = 0   # Real

# WELFake already has label column (just ensure correct format)
welfake["label"] = welfake["label"].astype(int)
```

```python
# Combine title + text
isot_fake["content"] = isot_fake["title"] + " " + isot_fake["text"]
isot_true["content"] = isot_true["title"] + " " + isot_true["text"]

welfake["content"] = welfake["title"] + " " + welfake["text"]
```

```python
isot_fake = isot_fake[["content", "label"]]
isot_true = isot_true[["content", "label"]]
welfake = welfake[["content", "label"]]
```

```python
isot_df = pd.concat([isot_fake, isot_true], ignore_index=True)

print("Final ISOT Shape:", isot_df.shape)
```

```python
# Fill missing values with empty string
isot_df["content"] = isot_df["content"].fillna("")
welfake["content"] = welfake["content"].fillna("")
```

```python
isot_df = pd.concat([isot_fake, isot_true], ignore_index=True)

print("Final ISOT Shape:", isot_df.shape)
print("WELFake Shape:", welfake.shape)
```

```python
import re
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = str(text).lower()
    
    # 🔥 REMOVE REUTERS BIAS (MOST IMPORTANT)
    text = re.sub(r'^.*?\(reuters\) - ', '', text)
    
    # remove special characters
    text = re.sub('[^a-zA-Z]', ' ', text)
    
    words = text.split()
    
    # remove stopwords
    words = [word for word in words if word not in stop_words]
    
    return " ".join(words)
```

```python
print("Cleaning ISOT dataset...")
isot_df["content"] = isot_df["content"].apply(clean_text)

print("Cleaning WELFake dataset...")
welfake["content"] = welfake["content"].apply(clean_text)
```

```python
# ISOT
isot_df.dropna(inplace=True)
isot_df.drop_duplicates(inplace=True)

# WELFake
welfake.dropna(inplace=True)
welfake.drop_duplicates(inplace=True)

print("ISOT after cleaning:", isot_df.shape)
print("WELFake after cleaning:", welfake.shape)
```

```python
# Save cleaned datasets
isot_df.to_csv("../Data/ISOT/isot_cleaned.csv", index=False)
welfake.to_csv("../Data/WELFake/welfake_cleaned.csv", index=False)

print("✅ Cleaned datasets saved successfully!")
```
