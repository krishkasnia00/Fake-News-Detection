```python
import pandas as pd
import os
import pickle

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
BASE_PATH = "../Data"

isot_path = os.path.join(BASE_PATH, "ISOT", "isot_cleaned.csv")
welfake_path = os.path.join(BASE_PATH, "WELFake", "welfake_cleaned.csv")

isot_df = pd.read_csv(isot_path)
welfake_df = pd.read_csv(welfake_path)

print("ISOT Shape:", isot_df.shape)
print("WELFake Shape:", welfake_df.shape)
```

```python
isot_df["content"] = isot_df["content"].astype(str)
welfake_df["content"] = welfake_df["content"].astype(str)
```

```python
# Fix missing values (VERY IMPORTANT)
isot_df["content"] = isot_df["content"].fillna("")
welfake_df["content"] = welfake_df["content"].fillna("")
```

```python
X_isot = isot_df["content"]
y_isot = isot_df["label"]

X_welfake = welfake_df["content"]
y_welfake = welfake_df["label"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Improved TF-IDF
tfidf_isot = TfidfVectorizer(
    max_features=10000,        # 🔥 increased
    ngram_range=(1,3),         # 🔥 unigram + bigram + trigram
    min_df=5,                  # ignore rare words
    max_df=0.7                 # ignore too common words
)

X_isot_tfidf = tfidf_isot.fit_transform(X_isot)

print("ISOT TF-IDF shape:", X_isot_tfidf.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Improved TF-IDF
tfidf_isot = TfidfVectorizer(
    max_features=10000,        # 🔥 increased
    ngram_range=(1,3),         # 🔥 unigram + bigram + trigram
    min_df=5,                  # ignore rare words
    max_df=0.7                 # ignore too common words
)

X_isot_tfidf = tfidf_isot.fit_transform(X_isot)

print("ISOT TF-IDF shape:", X_isot_tfidf.shape)
```

```python
# TF-IDF for WELFake
tfidf_welfake = TfidfVectorizer(
    max_features=10000,
    ngram_range=(1,3),
    min_df=5,
    max_df=0.7
)

X_welfake_tfidf = tfidf_welfake.fit_transform(X_welfake)

print("WELFake TF-IDF shape:", X_welfake_tfidf.shape)
```

```python
os.makedirs("../models", exist_ok=True)

pickle.dump(tfidf_isot, open("../models/tfidf_isot.pkl", "wb"))
pickle.dump(tfidf_welfake, open("../models/tfidf_welfake.pkl", "wb"))

print("✅ TF-IDF models saved")
```

```python
bow_isot = CountVectorizer(max_features=5000)

X_isot_bow = bow_isot.fit_transform(X_isot)

print("ISOT BoW shape:", X_isot_bow.shape)
```

```python
bow_welfake = CountVectorizer(max_features=5000)

X_welfake_bow = bow_welfake.fit_transform(X_welfake)

print("WELFake BoW shape:", X_welfake_bow.shape)
```

```python
pickle.dump(bow_isot, open("../models/bow_isot.pkl", "wb"))
pickle.dump(bow_welfake, open("../models/bow_welfake.pkl", "wb"))

print("✅ BoW models saved")
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# ISOT Tokenizer
tokenizer_isot = Tokenizer(num_words=5000)
tokenizer_isot.fit_on_texts(X_isot)

# WELFake Tokenizer
tokenizer_welfake = Tokenizer(num_words=5000)
tokenizer_welfake.fit_on_texts(X_welfake)
```

```python
X_isot_seq = tokenizer_isot.texts_to_sequences(X_isot)
X_welfake_seq = tokenizer_welfake.texts_to_sequences(X_welfake)
```

```python
X_isot_pad = pad_sequences(X_isot_seq, maxlen=200)
X_welfake_pad = pad_sequences(X_welfake_seq, maxlen=200)

print("ISOT padded shape:", X_isot_pad.shape)
print("WELFake padded shape:", X_welfake_pad.shape)
```

```python
pickle.dump(tokenizer_isot, open("../models/tokenizer_isot.pkl", "wb"))
pickle.dump(tokenizer_welfake, open("../models/tokenizer_welfake.pkl", "wb"))

print("✅ Tokenizers saved")
```
